## Data Preprocessing

```python
def rate_process(i):
    
    if (i =='NEW') or (i == '-') or (i == 'nan'):
        return np.nan 
    else:
        i = str(i).split('/')
        i = i[0]
        return float(i)
```

```python
df_train['rate'] = df_train['rate'].apply(rate_process)
```

Type 2: To remove commas from the integer value

```python
def comma_process(i):
    i = str(i)
    
    if ',' in i:
        i = i.replace(',','')
        return float(i)
    else:
        return float(i)
```

def comma_process will get a value from .apply fucntion. If the ',' comma is present in the value them replace that comma with noting and return its float value to the function. On the other hand, if commma is not present then return its float value only.

```python
df_train['cost2ppl'].apply(comma_process)
```

```python
df_train['cost2ppl']=df_train['cost2ppl'].apply(comma_process)
```
